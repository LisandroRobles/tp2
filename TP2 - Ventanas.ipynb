```python
#############################################################################
# Importación de paquetes necesarios para la realización del ejercicio # 1  #
#############################################################################

#Paquetes estándar

#Paquete numérico
import numpy as np

#Paquete gráfico
import matplotlib as mpl
import matplotlib.pyplot as plt

#Paquete para manejo de datos
from pandas import DataFrame
from IPython.display import HTML

#Paquetes propios

#Paquete que implementa la clase analizador de espectro
import pdsmodulos.spectrum_analyzer as sa

#Paquete que implemento las funciones generadoras de ventanas
import pdsmodulos.windows as win
```

```python
################################################
# Parámetros del muestreo para el ejercicio #1 #
################################################

#Cantidad de puntos de la ventana
N1 = np.power(2,10)     

#Cantidad de puntos de la fft (zero pad para ver el espectro)
N2 = np.power(2,20)   

#Frecuencia de muestreo
fs = np.power(2,10)   
```

```python
#############################################
# Inicialización del analizador de espectro #
#############################################

SA = sa.spectrum_analyzer(fs,N2)
```

```python
##############################
# Generación de las ventanas #
##############################

#Generación de una ventana Rectangular  
wr = win.rectangular(N1)

#Generación de una ventana Bartlett(triangular)    
wt = win.bartlett(N1)

#Generación de una ventana Hann   
whn = win.hann(N1)

#Generación de una ventana Blackman    
wb = win.blackman(N1)

#Generación de una ventana Flat-Top    
wft = win.flattop(N1)
```

```python
###############################################
# Generación de los espectros de las ventanas #
###############################################

#Cálculo del espectro de la ventana Rectangular (Kernel de Dirichlet)
(f,Wr) = SA.module(wr,db = True)
    
#Cálculo del espectro de la ventana Bartlett (Triangular)
(f,Wt) = SA.module(wt,db = True)

#Cálculo del espectro de la ventana de Hann
(f,Whn) = SA.module(whn,db = True)

#Cálculo del espectro de la ventana de Blackman
(f,Wb) = SA.module(wb,db = True)

#Calculo del espectro de la ventana Flat-Top
(f,Wft) = SA.module(wft,db = True)
```

```python
##########################################
# Presentación gráfica de los resultados #
##########################################

plt.figure()
fig, axarr = plt.subplots(5, 2,figsize = (10,20)) 
fig.suptitle('Visualización tiempo-frecuencia de las ventanas, N =' + str(N1),fontsize=12,y = 1.08)
fig.tight_layout()

axarr[0,0].plot(wr)
axarr[0,0].set_title('Representación temporal de la Ventana Rectangular')
axarr[0,0].set_ylabel('$w_{r}(t)$[V]')
axarr[0,0].set_xlabel('$n$')
axarr[0,0].axis('tight')
axarr[0,0].grid()

axarr[0,1].plot(f[0:int(N2/32)],Wr[0:int(N2/32),0])
axarr[0,1].set_title('Espectro de módulo de la Ventana Rectangular')
axarr[0,1].set_ylabel('$W_{r}(f)$[V/Hz]')
axarr[0,1].set_xlabel('$f$[Hz]')
axarr[0,1].axis('tight')
axarr[0,1].grid()

axarr[1,0].plot(wt)
axarr[1,0].set_title('Representación temporal de la Ventana de Bartlett')
axarr[1,0].set_ylabel('$w_{t}(t)$[V]')
axarr[1,0].set_xlabel('$n$')
axarr[1,0].axis('tight')
axarr[1,0].grid()

axarr[1,1].plot(f[0:int(N2/32)],Wt[0:int(N2/32),0])
axarr[1,1].set_title('Espectro de módulo de la Ventana de Bartlett')
axarr[1,1].set_ylabel('$W_{t}(f)$[V/Hz]')
axarr[1,1].set_xlabel('$f$[Hz]')
axarr[1,1].axis('tight')
axarr[1,1].grid()

axarr[2,0].plot(whn)
axarr[2,0].set_title('Representación temporal de la Ventana de Hanning')
axarr[2,0].set_ylabel('$w_{hn}(t)$[V]')
axarr[2,0].set_xlabel('$n$')
axarr[2,0].axis('tight')
axarr[2,0].grid()

axarr[2,1].plot(f[0:int(N2/32)],Whn[0:int(N2/32),0])
axarr[2,1].set_title('Espectro de módulo de la Ventana de Hanning')
axarr[2,1].set_ylabel('$W_{hn}(f)$[V/Hz]')
axarr[2,1].set_xlabel('$f$[Hz]')
axarr[2,1].axis('tight')
axarr[2,1].grid()

axarr[3,0].plot(wb)
axarr[3,0].set_title('Representación temporal de la Ventana de Blackman')
axarr[3,0].set_ylabel('$w_{b}(t)$[V]')
axarr[3,0].set_xlabel('$n$')
axarr[3,0].axis('tight')
axarr[3,0].grid()

axarr[3,1].plot(f[0:int(N2/32)],Wb[0:int(N2/32),0])
axarr[3,1].set_title('Espectro de módulo de la Ventana de Blackman')
axarr[3,1].set_ylabel('$W_{b}(f)$[V/Hz]')
axarr[3,1].set_xlabel('$f$[Hz]')
axarr[3,1].axis('tight')
axarr[3,1].grid()

axarr[4,0].plot(wft)
axarr[4,0].set_title('Representación temporal de la Ventana Flat-Top')
axarr[4,0].set_ylabel('$w_{ft}(t)$[V]')
axarr[4,0].set_xlabel('$n$')
axarr[4,0].axis('tight')
axarr[4,0].grid()

axarr[4,1].plot(f[0:int(N2/32)],Wft[0:int(N2/32),0])
axarr[4,1].set_title('Espectro de módulo de la Ventana Flato-Top')
axarr[4,1].set_ylabel('$W_{ft}(f)$[V/Hz]')
axarr[4,1].set_xlabel('$f$[Hz]')
axarr[4,1].axis('tight')
axarr[4,1].grid()
```

```python
############################################################
# Presentación gráfica de los resultados todos en uno solo #
############################################################

plt.figure()
fig, axarr = plt.subplots(1, 1,figsize = (10,5)) 
fig.suptitle('Comparación espectral de las ventanas, N =' + str(N1),fontsize=12,y = 1.08)
fig.tight_layout()

axarr.plot(f[0:int(N2/32)],Wr[0:int(N2/32),0],f[0:int(N2/32)],Wt[0:int(N2/32),0],f[0:int(N2/32)],Whn[0:int(N2/32),0],f[0:int(N2/32)],Wb[0:int(N2/32),0],f[0:int(N2/32)],Wft[0:int(N2/32),0])
#axarr[0].set_title('Representación temporal de la Ventana Rectangular')
axarr.set_ylabel('$W(f)$[V]')
axarr.set_xlabel('$f$[Hz]')
axarr.legend(['Rectangular','Bartlett','Hanninb','Blackman','Flat-Top'],loc = 'upper right')
axarr.axis('tight')
axarr.grid()
```
